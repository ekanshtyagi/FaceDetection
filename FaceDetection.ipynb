```python
#I have used cv2_imshow instead of cv2.imshow as the latter is banned in GOOGLE COLAB and the cv2_imshow needs to be imported before usage
from google.colab.patches import cv2_imshow
```

```python
#import the libraries
import numpy
import cv2
import matplotlib.pyplot as plt
```

```python
#importing the image and training model
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
image = cv2.imread('/content/modern-family-og-min.webp')

#turn the image into grayscale
grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#detecting the faces
faces = face_cascade.detectMultiScale(grayImage , minNeighbors=5 , scaleFactor=1.08)
```

```python
print(type(faces)) 
#imread returns a numpy array if image path is given, the array returns the coordinates of each and every face

#printing the number of faces found
if len(faces) == 0:
    print("No faces found")
  
else:
    print(faces) #array
    print(faces.shape) #dimensions of the array(rows,columns); according to dimension of all faces
    print("Number of faces detected: " + str(faces.shape[0])) # 0 implies rows of the matrix i.e. no of faces
  
    for (x,y,w,h) in faces:
        #creating rectangle around each face
        cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,188),2) #(image, start_point, end_point, colour, border_radius)

        #creating seprate rectangle to display desired text
        cv2.rectangle(image, ((0,image.shape[0] -30)),(300, image.shape[0]), (255,255,255),-1)
        cv2.putText(image, " Number of faces detected: " + str(faces.shape[0]), (0,image.shape[0] -10), cv2.FONT_HERSHEY_SIMPLEX, 0.57,  (0,0,0), 2)
        #(image, text,org(Position of text),font,font_size, colour of text, thickness of text)

        plt.imshow(image)
```

```python
#display the output
cv2_imshow(image)
```
